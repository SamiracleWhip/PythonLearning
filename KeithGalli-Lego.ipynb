```python
import pandas as pd
import numpy as np
```

```python
url = 'https://github.com/KeithGalli/lego-analysis/blob/master/datasets/lego_sets.csv?raw=true'
df = pd.read_csv(url)
```

```python
parent_theme = pd.read_csv(r"https://github.com/KeithGalli/lego-analysis/blob/master/datasets/parent_themes.csv?raw=true")
```

```python
df.head()
```

```python
parent_theme.head()
```

```python
#QUESTION 1: What % of all licensed sets ever released were Star Wars themed?

#FIRST STEP: We want to join the parent them with the name
```

```python
merged= df.merge(parent_theme, left_on="parent_theme",right_on="name")
```

```python
merged.drop(columns="name_y", inplace=True) #Remove repeated/redundant columns
merged.head()
```

```python
licensed = merged[merged["is_licensed"]] #Filters only True values- is same as doing "=True"
```

```python
licensed.head() #only looks at licensed products now
```

```python
star_wars = licensed[licensed["parent_theme"]=="Star Wars"]

star_wars.head() #Further filters by Licensed AND Star Wars
```

```python
#Check and filter out the Null columns
merged.head()
```

```python
merged[merged["set_num"].isnull()] #This is a kind of filter
```

```python
merged[merged["set_num"].isnull()].shape #153 Nulls need to be taken out
```

```python
licensed = licensed.dropna(subset = ["set_num"]) #Only drop the NAs from set_num
```

```python
licensed.shape
```

```python
star_wars.shape
```

```python
the_force=star_wars.shape[0]/licensed.shape[0]
```

```python
the_force
```

```python
#ANSWER: 51% of all sets were Starwars themed
```

```python
#PART 2: In which year was Star Wars NOT the most popular LICENSED theme?
#In terms of the number of sets released that year
```

```python
#You want licensed= True here so you can reuse the licensed df
```

```python
licensed.head()
```

```python
license_sorted = licensed.sort_values("year")
```

```python
license_sorted
```

```python
summed_df=license_sorted.groupby(["year","parent_theme"]).sum().reset_index()
```

```python
summed_df.head() #is_licensed will also be summed because it is a BOOLEAN and hence True = 1 :)
```

```python
#We want to get the max row of anything that shares the same year
```

```python
#So you can sort and then drop duplicates below!
```

```python
max_df = summed_df.sort_values("is_licensed",ascending=False).drop_duplicates(["year"])
```

```python
max_df.sort_values("year",inplace=True)
```

```python
max_df
```

```python
# So 2017 is the answer because it is NOT StarWars
```

```python
# Question 3: Which year had the most number of Sets?
```

```python
clean_df = merged[~merged["set_num"].isnull()] #the tilda means anything but
```

```python
clean_df["count"] = 1
clean_df.head()
```

```python
sets_per_year = clean_df.groupby(["year"]).sum().reset_index()[["year","count"]]
```

```python
sets_per_year
```
